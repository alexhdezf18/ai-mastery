```python
import pandas as pd

datos_ventas = {
    'Producto': ["Laptop", "Mouse", "Monitor", "Teclado", "Cable HDMI"],
    'Precio': [800, 20, 200, 50, 10],
    'Cantidad': [5, 50, 10, 20, 100]
}

df = pd.DataFrame(datos_ventas)
print(df)
```

```python
df['Total'] = df['Precio'] * df['Cantidad']
print(df)
```

```python
productos_top = df[ df['Total'] > 500 ]
print(productos_top)
```

```python
ingreso_total = df['Total'].sum()
print(f"Ingresos de hoy: ${ingreso_total}")
```

```python
df.to_csv('reporte_ventas.csv', index=False)

print("¡Archivo guardado con éxito!")
```
