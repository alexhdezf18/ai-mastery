```python
import pandas as pd
import matplotlib.pyplot as plt

datos_ventas = {
    'Producto': ["Laptop", "Mouse", "Monitor", "Teclado", "Cable HDMI"],
    'Precio': [800, 20, 200, 50, 10],
    'Cantidad': [5, 50, 10, 20, 100]
}

df = pd.DataFrame(datos_ventas)
df['Total'] = df['Precio'] * df['Cantidad']
```

```python
plt.figure(figsize=(10, 6))
plt.bar(df['Producto'], df['Total'], color='red')
plt.title("Ventas Totales por Producto")
plt.xlabel("Productos")
plt.ylabel("Dinero (USD)")
plt.savefig('grafica_ventas.png')
#plt.show()
```
