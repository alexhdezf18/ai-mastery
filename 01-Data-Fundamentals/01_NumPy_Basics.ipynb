```python
import numpy as np

# Generamos 20 calificaciones simuladas entre 50 y 100
# np.random.seed(42) sirve para que siempre salgan los mismos números (útil para aprender)
np.random.seed(42)
calificaciones = np.random.randint(50, 101, 20)

print("Todas las calificaciones:", calificaciones)
```

```python
# Queremos calificaciones MAYORES a 80 Y MENORES a 95
# En IA, el "Y" se escribe con "&" y cada condición va en paréntesis ()

filtro = (calificaciones > 80) & (calificaciones < 95)

# Aplicamos el filtro a la lista original
resultado = calificaciones[filtro]

print("Alumnos destacados (81-94):", resultado)
```
