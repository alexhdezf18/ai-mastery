```python
from sklearn.model_selection import train_test_split
```

```python
import pandas as pd
import numpy as np

np.random.seed(42)
metros = np.random.randint(60, 300, 20)
precios = (metros * 10) + np.random.randint(1000, 5000, 20)
df = pd.DataFrame({'Metros': metros, 'Precio': precios})

x = df[['Metros']]
y = df['Precio']

print(f"Tenemos {len(x)} casas listas.")
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
print(len(X_train))
```

```python
from sklearn.linear_model import LinearRegression

modelo = LinearRegression()
modelo.fit(X_train, y_train)
```

```python
y_pred = modelo.predict(X_test)

comparativa = pd.DataFrame({'Real': y_test, 'Predicho:': y_pred})
print(comparativa)
```

```python
from sklearn.metrics import mean_absolute_error

error = mean_absolute_error(y_test, y_pred)
print(error)
```

```python
from matplotlib import pyplot as plt

plt.scatter(X_train, y_train, color='blue', label='Entrenamiento')
plt.scatter(X_test, y_test, color='red', label='Prueba')

plt.plot(X_test, y_pred, color="black", linewidth=3)
plt.legend()
plt.show()
```
