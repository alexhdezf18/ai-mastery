```python
import pandas as pd

data = {
    'Metros': [80, 100, 120, 150, 200],
    'Cuartos': [2, 3, 3, 4, 5],
    'Precio': [180, 220, 260, 320, 420]
}

df = pd.DataFrame(data)
print(df)
```

```python
from sklearn.linear_model import LinearRegression

x = df[['Metros', 'Cuartos']]
y = df['Precio']

modelo = LinearRegression()
modelo.fit(x, y)

print(modelo.coef_)
```

```python
print(modelo.predict([[130, 3]]))
```
