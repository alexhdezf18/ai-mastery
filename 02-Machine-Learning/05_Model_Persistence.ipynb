```python
import pandas as pd
from sklearn.linear_model import LinearRegression
import joblib
```

```python
data = {
    'x' : [1, 2, 3, 4, 5],
    'y' : [10, 20, 30, 40, 50]
}
df = pd.DataFrame(data)

X = df[['x']]
y = df['y']
```

```python
model = LinearRegression()
model.fit(X, y)

print(model.coef_)
```

```python
joblib.dump(model, 'modelo_prueba.joblib')
```
